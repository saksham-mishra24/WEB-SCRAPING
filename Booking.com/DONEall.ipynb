```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
hotel_details = []
```

```python
content = soup.find_all('div', {'data-testid': "property-card"})
```

```python
for hotels in content:
   name = hotels.find('div',{'data-testid':"title"}).text#.text gives only hotel names
   reviews = hotels.find('div',{'class':"d8eab2cf7f c90c0a70d3 db63693c62"}).text.strip()
   scored = hotels.find('div',{'class':"b5cd09854e d10a6220b4"}).text.strip()
   address = hotels.find('span',{'data-testid':"address"}).text.strip()
   distance = hotels.find('span',{'data-testid':"distance"}).text.strip()

   hotel_info = {'Hotel':name,'Ratings':scored,'No_of_Reviews':reviews,'address':address, 'distance':distance}
   hotel_details.append(hotel_info)
print(hotel_details)
```
